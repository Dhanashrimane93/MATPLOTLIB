```python
# Q plot count of genre 2 on a line chart where count is >1

q1=df['genre_2'].value_counts()
q1=q1[q1>1]
q1
```

```python
plt.figure(figsize=(9,4))
plt.plot(q1.index,q1.values,color='maroon',marker='o',
         label='count',markersize=12,lw=3,linestyle=':')
plt.xlabel('genre2')
plt.ylabel('count')
plt.legend()
plt.title('genre2 vs count',fontsize=15,color='blue')
plt.grid()
plt.show()
```

```python
# Q PLOT YEAR WISE PROFIT ON A LINE CHART WHERE PROFIT IS GROSS BUDGET

d1["Gross"]=d1["Gross"]/1000000
d1["budget"]=d1["budget"]/1000000
d1.head()
```

```python
d1['profit']=d1['Gross']-d1['budget']
d1.head()
```

```python
q2=d1.groupby('title_year')['profit'].mean().reset_index()
q2=q2.sort_values(by='title_year',ascending=True)
q2
```

```python
plt.plot(q2['title_year'],q2['profit'],color='brown',marker='*')
plt.title('year wise avg profit(in M)',fontsize=13,color='blue')
plt.xlabel('title_year')
plt.ylabel('profit')
plt.grid()
```

```python
q3=d1.groupby('title_year')['budget','profit'].mean().reset_index()
q3=q3.sort_values(by='title_year',ascending=True)
q3
```

```python
plt.plot(q3['title_year'],q3['budget'],color='green',marker='*',label='budget')
plt.plot(q3['title_year'],q3['profit'],color='brown',marker='*',label='profit')
plt.title('year wise avg budget and profit(in M)',fontsize=30,color='blue')
plt.xlabel('tilte_year')
plt.legend()
plt.grid()
plt.show()
```
