```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('DataScience_Marks_Dataset.csv')
```

```python
# Q.1 create a simple line plot of namevs total_marks
plt.plot(df['Name'],df['Total_marks'])
```

```python
# Q .2 create a bar plot of name vs SQL and apply the following customization to the plot

plt.figure(figsize=(9,4))
plt.plot(df['Name'],df['Python'],color='indigo',marker='o',
        label='Python',markersize=10,lw=2,linestyle='-.')
plt.xlabel('name')
plt.ylabel('python')
plt.legend()
plt.title('Name vs python',fontsize=15,color='purple')
plt.grid() # use to show grid lines in graph
plt.show()
```

```python
# Q.3 create a line plot of name vs python,statistics and SQL ansd apply
#the folowing customization to the plot.

plt.figure(figsize=(9,4))
plt.plot(df['Name'],df['Python'],color='cyan',marker='^',label='python',markersize=10)
plt.plot(df['Name'],df['Statistics'],color='red',marker='o',label='Statistics',markersize=10)
plt.plot(df['Name'],df['SQL'],color='purple',marker='*',label='SQL',markersize=10)
plt.title('Name wise python ,Statistics and SQL',fontsize=13,color='blue')
plt.xlabel('Name')
plt.legend()
plt.grid()
plt.show()
```

```python
# Q.4 create a scatter plot of python vs total marks and apply the following customization

plt.figure(figsize=(9,4))
plt.scatter(df['Python'],df['Total_marks'],marker='h',s=40,color='orange')
plt.xlabel('Python')
plt.ylabel('Total_marks')
plt.title('Python vs total_marks',fontsize=15,color='purple')
plt.grid()
plt.show()
```

```python
# Q.5 create a scatter plot of name vs python statistics and SQL in the same plot and apply the  following customization

plt.figure(figsize=(9,4))
plt.scatter(df['Name'],df['Python'],marker='s',s=45,color='teal',label='Name vs python')
plt.scatter(df['Name'],df['Statistics'],marker='p',s=45,color='salmon',label='Name vs statistics')
plt.scatter(df['Name'],df['SQL'],marker='h',s=45,color='orchid',label='Name vs SQL')
plt.title(' Name vs Python,statistics and SQL',fontsize=15,color='black')
plt.grid()
plt.legend()
plt.show()
```
