```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('DataScience_Marks_Dataset.csv')
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,figsize=(10,12))
ax1.bar(df['Name'],df['SQL'],color='orange',label='SQL',edgecolor='black')
ax1.set_title('Name wise SQL marks',fontsize=15,color='blue')
ax1.set_xlabel('Name')
ax1.legend()

ax2.plot(df['Name'],df['Python'],color='cyan',marker='^',label='python',markersize=10)
ax2.plot(df['Name'],df['Statistics'],color='red',marker='o',label='Statistics',markersize=10)
ax2.plot(df['Name'],df['SQL'],color='purple',marker='*',label='SQL',markersize=10)
ax2.set_title('Name wise  python sql and statistics',fontsize=10)
ax2.set_xlabel('Name')
ax2.set_ylabel('Marks')
ax2.legend()
ax2.grid()

ax3.pie(x=df['Total_marks'],labels=df['Name'],
        colors=['red','blue','yellow','green','cyan','deeppink','dodgerblue','orange'],autopct='%.2f%%',
        explode=(0,0,0,0,0,0.2,0,0.1),shadow=True,counterclock=False)
ax3.legend(df['Name'],title='Names',loc='right',bbox_to_anchor=(0.9,1,0.6,-0.5))

ax4.stackplot(df['Name'],df['Python'],df['SQL'],df['Statistics'],labels=['python','SQL','statistics'])
ax4.set_xlabel('Name')
ax4.set_ylabel('marks')
ax4.set_ylim(0,400)
ax4.legend()
ax4.set_title('Name wise python sql statistics marks',fontsize=15,color='blue')

plt.show()
```
