```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('DataScience_Marks_Dataset.csv')
```

```python
# Q.1 Create a simple bar plot of name vs total marks
plt.bar(df['Name'],df['Total_marks'])
```

```python
#Q 2. create a bar plot of name  vs SQL and apply the following customization to the plot

#make the fig size
plt.figure(figsize=(10,8))
plt.bar(df['Name'],df['SQL'],color='violet',edgecolor='black')
plt.title('Name wise SQL',fontsize=15,color='red')
plt.xlabel('Name')
plt.ylabel('SQL')
plt.xticks(rotation=90)
plt.savefig('Name_vs_SQL.png')
plt.show()
```

```python
# Q.3 CREATE A BAR PLOT OF NAME VS PYTHON  AND APPLT THE FOLLOWING 

plt.figure(figsize=(10,8))
plt.bar(df['Name'],df['Python'],color=['orange','lightblue','violet','lightgreen','brown','tan','yellow','pink'],edgecolor='black')
plt.title('Name wise python',fontsize=15,color='red')
plt.xlabel('Name')
plt.ylabel('Python')
plt.xticks(rotation=60)
plt.show()
```

```python
# Q 4. CREATE A BAR PLOT OF NAME  VS STATSTICS AND APPLY THE FOLLOWING 

s=list(range(8))
plt.figure(figsize=(10,8))
plt.bar(s,df['Statistics'],color='cyan',label='Statistics',edgecolor='black')
plt.ylim(0,120)
for idx,val in enumerate(df['Statistics']):
    plt.annotate(xy=[s[idx]-0.2,int(val)+2],text=int(val))
    plt.xticks(s,df['Name'],rotation=90)
    plt.title('Name wise Statistics',fontsize=15,color='red')
    plt.xlabel('Name')
    plt.ylabel('Python')
    plt.legend()
    plt.show()
```
